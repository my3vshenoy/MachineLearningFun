```python
# Hot and cold Method
# Step 1: An empty network
weight = 0.1
lr = 0.01
def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
# Step 2
# Predict and get the raw error
number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0] # 8.5
true = win_or_lose_binary[0] #1

prediction = neural_network(input, weight)

## MEAN SQUARED ERROR: One method of measuring error
error = (prediction - true) ** 2
print(error) # raw error
```

```python
# Step 3: Compare with a higher weight to get the new raw error
lr = 0.01
prediction_with_higher_lr = neural_network(input, weight + lr)
print(prediction_with_higher_lr)

error_with_higher_lr = (prediction_with_higher_lr - true) ** 2
print(error_with_higher_lr) # raw error
```

```python
# Step 4: Compare with a lower weight to get the new raw error
lr = 0.01
prediction_with_lower_lr = neural_network(input, weight - lr)
print(prediction_with_lower_lr)

error_with_lower_lr = (prediction_with_lower_lr - true) ** 2
print(error_with_lower_lr) # raw error
```

```python
# Step 5: LEARN Compare the errors and set the new weight
# Increased Weight Error < Decreased Weight Error
# Hence new weight: 0.11
# This is one iteration of the hot and cold method.
# We need to run multiple of them to get the error == 0
```

```python
# Hot & Cold Learning Difficult Problem
knob_weight = 0.5
input = 0.5
goal_prediction = 0.8


lr = 0.001

for iteration in range(2003): # Repeat learnings 2001 times 
    prediction = input * weight
    error = (prediction - goal_pred) ** 2
    
    print("Error: " + str(error) + "  Prediction: " + str(prediction))
    
    up_prediction = input * (weight + lr)
    up_error = (up_prediction - goal_prediction) ** 2
    
    down_prediction = input * (weight - lr)
    down_error = (down_prediction - goal_prediction) ** 2
    
    if (up_error > down_error): 
        weight = weight - lr
    
    if (up_error < down_error):
        weight = weight + lr
    
print("Final weight: " + str(weight))
# Final weight: 1.2009999999999785 
    
```

```python
# One iteration of gradient descendant with alpha
# Step 3: Calculate the node delta and putting it on the output node

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0] # 8.5
goal_prediction = win_or_lose_binary[0] #1

prediction = neural_network(input, weight)

## MEAN SQUARED ERROR
error = (prediction - goal_prediction) ** 2
print(error) # raw error

# Pure error or Node delta: Measurement of how much this node missed
delta_error = prediction - goal_prediction
print(delta_error)
```

```python
# One iteration of gradient descendant with alpha
# Step 4: Use delta_error to calculate direction_and_amount, here: weight_delta

weight_delta = input * delta_error # Stopping, negative reversal, scaling
print(weight_delta)
```
